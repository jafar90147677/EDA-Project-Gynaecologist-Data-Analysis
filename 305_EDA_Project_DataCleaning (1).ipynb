## **Filling Null Values with `0`,`Unknown` and `Not Mentioned` and Data Type Conversion**

```python
# filling experience null values with Not Mentioned
df['Experience (in years)']=df['Experience (in years)'].fillna(0)
```

```python
# filling education null values with Not Mentioned
df['Education']=df['Education'].fillna('Not Mentioned')
```

### Conversion of `Fee at clinic` column from object to numeric

- Removing `₹ ` from the columns so the column can be converted to numerical type
- Converting datatype of the column using `pd.to_numeric() method`

```python
# replacing clinic fee rupees symbol with space
df['Fee at clinic']=df['Fee at clinic'].replace('₹',' ') 

# Converting categorical to numerical
df['Fee at clinic']=pd.to_numeric(df['Fee at clinic'])

# filling fee at clinic null values with unknown
df['Fee at clinic']=df['Fee at clinic'].fillna(0)
```

## **Conversion of `Online Fee` column from object to numeric**

- Removing `₹ ` from the columns so the column can be converted to numerical type
- Converting datatype of the column using `pd.to_numeric() method`

```python
# replacing online fee rupees symbol with space
df['Online Fee']=df['Online Fee'].replace('₹',' ') 

# Converting categorical to numerical
df['Online Fee']=pd.to_numeric(df['Online Fee'])

# filling online fee null values with 'zero'
df['Online Fee']=df['Online Fee'].fillna(300)
```

```python
# filling clinic name null values with unknown
df['Clinic Name']=df['Clinic Name'].fillna('Unknown')
```

```python
# filling Area null values with unknown
df['Area']=df['Area'].fillna('Unknown')
```

## **Adding new column Total Fee**
- using sum of `Online Fee` and `Fee at clinic` columns

```python
# Adding new Column
df['Total Fee'] = df['Fee at clinic'] + df['Online Fee']

# Format Total Fee as currency
df['Total Fee'] = '₹ ' + df['Total Fee'].astype(int).astype(str)

# replacing Total fee rupees symbol with space
df['Total Fee']=df['Total Fee'].str.replace('₹',' ') 
df['Total Fee']=pd.to_numeric(df['Total Fee'])
```

## **converting Timing columns to datetime object**
1. Converting `Available From` and `Available Till` columns to datetime format.


```python
df['Available From'] = pd.to_datetime(df['Available From'], format='%I:%M %p').dt.time
df['Available Till'] = pd.to_datetime(df['Available Till'], format='%I:%M %p').dt.time
```

## **Creating New Column Number of Hrs Available**

1. Creating new column `Number of Hrs. Available` to show total number of hours of availability by subtracting time delta of `Available Till`  from time delta of `Available From`
2. Removing days part from the `Number of Hrs. Available` column

```python
# Adding new column i.e., Adding Numbers of Hrs.Available
df['Number of Hrs. Available'] = (pd.to_timedelta(df['Available Till'].astype(str)) - pd.to_timedelta(df['Available From'].astype(str)))
```

```python
hrs = []
for i in df['Number of Hrs. Available'] :
    b = re.findall(r'days (.+):00$',str(i))
    if len(b) > 0 :
        if b[0][0] == '+' :
            hrs.append(b[0][1:])
        else :
            hrs.append(b[0])
    else :
        hrs.append(i)
```

```python
df['Number of Hrs. Available'] = hrs
```

```python
df.rename({'Number of Hrs. Available' : 'Number of Hrs. Available (HH:MM)'},inplace = True)
```

```python
twelve_hr = df[pd.to_datetime(df['Number of Hrs. Available']).dt.hour <= 12]
```
